```python
import pandas as pd
import numpy as np
```

```python
# Creamos listas de Grados, y luego un diccionario
students = [ "Gissela", "Daniel", "Andres", "Sandra", "Rosalyn" ]
math     = [ 16, 14, 17, 17, 17 ]
english  = [ 16, 17, 19, 18, 15 ]
art      = [ 11, 17, 13, 14, 17 ]

# Dictionary
diplomado = {'Students':students, 'Math':math, 'English':english, 'Art':art}
gradesA1 = pd.DataFrame( diplomado)
gradesA1
```

```python
#Empezamos a ordenar la base de datos según sus valores. Lo que hace Phyton es mostrarnos una ventana de cómo sería la BD ordenada. Sin embargo no modifica la BD original
gradesA1.sort_values("Students")
```

```python
#Para ello, usamos la función inplace
gradesA1.sort_values("Students", inplace=True)
gradesA1
```

```python
#Ahora vamos a crear una columna con el promedio de los almunos. Haremos operaciones matemáticas
gradesA1["promedio"] = (gradesA1["Math"]+gradesA1["English"]+gradesA1["Art"]) / 3
gradesA1
```

```python
#Ahora le pido a Phyton que elimine columnas. Para ello no solo tengo que señalarle el nombre de la columna, sino donde esta ubicado. Para ello, axis es el eje. El eje Y (horizontal, las filas) = 0, y el eje X (vertical, las columnas) = 1
gradesA1.drop("promedio", axis=1, inplace=False)
```

# Aplicamos a un caso real de manejo de base de datos

```python
#Ahora vamos a leer bases de datos de GitHub: https://github.com/github/innovationgraph/tree/main/data
```

```python
#Vamos a elaborar un análisis de datos unicamente a nivel de países de LATAM
#Obtenemos la data de github

import pandas as pd

url = "https://raw.githubusercontent.com/github/innovationgraph/refs/heads/main/data/repositories.csv"

#Carga directa del URL

df = pd.read_csv(url)
df
```

```python
## Renombremos de manera correcta las columnas

df.rename(columns={"repositories":"num_repos"}, inplace=True)
df
```

```python
## Ahora seleccionamos valores solo para LATAM
#Para eso, pongamos de ejemplo la selección de casos donde solo está PERÚ
df[df["iso2_code"]=="PE"]
#Entonces, para seleccionar debes seleccionar la columna, y señalar que dato buscamos
```

```python
#Ahora solo para LATAM
# List of South American ISO2 codes
south_america_codes = [
    "AR", "BO", "BR", "CL", "CO", "EC",
    "GY", "PY", "PE", "SR", "UY", "VE", "GF"  # GF optional (French Guiana)
]

# Filter DataFrame, y creamos el objeto "df_south_america"
df_south_america = df[df["iso2_code"].isin(south_america_codes)]
df_south_america
```

```python
#Ahora, quiero saber cuantas observaciones tengo. Para eso usamos shape (observaciones, columnas)
df_south_america.shape
#Para ver todas mis columnas
df_south_america. columns
```

# Ahora vamos a ordenar la BD por columnas

```python
# Reorder columns
df_south_america = df_south_america[["iso2_code", "year", "quarter", "num_repos"]]
df_south_america

#Aquí no estamos ordenando por sort.values, sino ordenando por indexación
```

```python
#Corregimos la indexación para ordenarlo de 0 a 4672
df_south_america.reset_index(inplace=True)
df_south_america

#Phyton me genero una columna index nueva, porque así me guarda la indexación pasada. Podemos borrarla, pero la dejamos
```

```python
print(df_south_america.columns.tolist())
```

```python
#Ahora ordenamos de manera ascendente. Eso, por el TRUE
df_south_america = df_south_america.sort_values(
    by=["iso2_code", "year", "quarter"],
    ascending=[True, True, True]
)
df_south_america
```

```python
df_south_america.drop("index", axis=1, inplace=True)
df_south_america
```

```python
#Multiplicamos 13*6*4 = 312
#La base de datos no esta balanceada. Esto implica que hay filas que no poseen datos completos
#Sabemos que la base de datos posee 273 observaciones, entonces hay un problema de balanceo
#Debemos entonces, conseguir las 312 observaciones
```

```python
import pandas as pd

# Ensure correct order and types
df_south_america = df_south_america[["iso2_code", "year", "quarter", "num_repos"]].copy()
df_south_america["year"] = df_south_america["year"].astype(int)
df_south_america["quarter"] = df_south_america["quarter"].astype(int)

# Define the full cartesian product (all combos)
all_iso2   = sorted(df_south_america["iso2_code"].unique())
all_years  = sorted(df_south_america["year"].unique())
all_quarts = [1, 2, 3, 4]   # or: sorted(df_south_america["quarter"].unique())

full_index = pd.MultiIndex.from_product(
    [all_iso2, all_years, all_quarts],
    names=["iso2_code", "year", "quarter"]
)

# Reindex to the full panel
balanced = (
    df_south_america
    .set_index(["iso2_code", "year", "quarter"])
    .reindex(full_index)
    .reset_index()
)
balanced
```

```python
# Optional: fill missing outcomes (choose ONE)
# If you'd rather keep them as NaN (to impute later), skip the fill.

# Final order & sort
balanced = balanced.sort_values(["iso2_code", "year", "quarter"]).reset_index(drop=True)

# Quick checks
print(balanced.shape)            # should be (312, 4)
print(balanced["num_repos"].isna().sum(), "missing cells created")

# If you want to see which combos were missing:
missing = balanced[balanced["num_repos"].isna()][["iso2_code", "year", "quarter"]]
# print(missing.head())

balanced
```

```python
# If "missing means zero"
balanced["num_repos"] = balanced["num_repos"].fillna(0).astype(int)
balanced
```

```python
#Ahora nos piden importar este csv para que lo trabajen
balanced.to_csv("../_data/github_repos_data_2020_2025_sa.csv")
```
