# Los "Tipos" en Phyton

Los tipos de datos (types) en Python son categorías que definen qué tipo de valor puede almacenar una variable y qué operaciones se pueden realizar con ese valor. Python es un lenguaje de tipado dinámico, lo que significa que no necesitas declarar explícitamente el tipo de una variable.

## Tipos básicos principales:

### Numéricos:

* int - Números enteros (ej: 5, -10, 0)
* float - Números decimales (ej: 3.14, -2.5)
* complex - Números complejos (ej: 3+4j)

### Texto:

str - Cadenas de texto (ej: "Hola", 'Python')

Puede ser con una comilla o dos. Para casos donde querramos poner como texto algo que esta inserto entre dos comillas, tendremos que posicionar una comilla ''.

### Booleanos:

bool - Valores lógicos (True o False)

Esto nos servirá cuando trabajemos con columnas, donde se registre en la memoria de Phyton que, por ejemplo, 1= Mujer y 0 = Varón
En Phyton:
* True = 1
* False = 0

### Secuencias:

* list - Listas mutables (ej: [1, 2, 3])
* tuple - Tuplas inmutables (ej: (1, 2, 3))
* range - Rangos de números

### Mapeo:

dict - Diccionarios (ej: {'nombre': 'Juan', 'edad': 25})

### Conjuntos:

* set - Conjuntos mutables
* frozenset - Conjuntos inmutables

### Otros:

NoneType - Representa la ausencia de valor (None)

¿Para qué sirven?
Los tipos de datos sirven para:

Organizar información de manera lógica
Determinar qué operaciones puedes hacer (no puedes sumar un texto con un número directamente)
Optimizar memoria y rendimiento
Prevenir errores en el código
Facilitar la comprensión del código
