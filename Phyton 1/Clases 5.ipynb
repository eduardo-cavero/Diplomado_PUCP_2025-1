# If y Loops
Los if (condicionales) y loops (bucles) en Python, que son estructuras fundamentales para controlar el flujo del programa.
1. IF (Condicionales)Los if permiten ejecutar código solo cuando se cumple una condición

Los loops permiten repetir código múltiples veces:
1. FOR Loop
Se usa cuando sabes cuántas veces quieres repetir algo

2. WHILE Loop
Se usa cuando no sabes exactamente cuántas veces repetir, pero tienes una condición

```python
import numpy as np
import pandas as pd
```

```python
pedro_grad = 2016
eduardo_grad = 2023
```

```python
#Usamos : para que un código este sujeto a la primera línea

if pedro_grad < eduardo_grad:
    print("Pedro y Eduardo no son de la misma promoción")
    print(5*5)

#No recibimos ningun output porque la información es falsa
#Si ponemos información cierta, si nos corre el código y tenemos output
```

```python
#Cuando Phyton reconoce que hay dos expresiones que son ciertas, solo va a correr el primer código de acuerdo al orden lógico

if pedro_grad == eduardo_grad:
    print("Pedro y Eduardo son de la misma promoción")

elif pedro_grad < eduardo_grad:
    print("Pedro y Eduardo no son de la misma promoción")

elif pedro_grad < eduardo_grad:
    print("Pedro y Eduardo no compartieron clases en algun momento")
```

```python
#OTRO EJEMPLO
country = "PE"
total = 75
#country = "AR"

if country == "PE":
    if total <= 50:
        print("Shipping Cost is  $50")
    elif total <= 100  and total>50:
        print("Shipping Cost is $25")
    elif total <= 150 and total>100:
        print("Shipping Costs $5")
    else:
        print("FREE")
        
elif country == "DE":
    if total <= 100:
        print("Shipping Cost is  $100")
    else:
        print("FREE")
    
else:
    if total <= 50:
        print("Shipping Cost is  $100")
    else:
        print("FREE")
```

```python
#LOOPS
sequence_ages = np.array( [15, 20, 25, 30 , 35 , 40 , 50 ] )
sequence_ages
```

```python
#Los Loops nos ayudará que un código permita que cada elemento de una lista sea aplicado
for m in sequence_ages:
    print(m-10)
```

```python
#Range en Python es una función que genera secuencias de números. 
#Es muy útil para crear rangos numéricos y se usa frecuentemente con loops.
for i in range(5):
    print(i)  # Genera: 0, 1, 2, 3, 4
```

```python
for age in range(6):
    print(sequence_ages[age+1] - sequence_ages[age])
```

El loop ejecuta 6 iteraciones (range(6) = 0, 1, 2, 3, 4, 5):

age = 0: sequence_ages[1] - sequence_ages[0] = 20 - 15 = 5
age = 1: sequence_ages[2] - sequence_ages[1] = 25 - 20 = 5
age = 2: sequence_ages[3] - sequence_ages[2] = 30 - 25 = 5
age = 3: sequence_ages[4] - sequence_ages[3] = 35 - 30 = 5
age = 4: sequence_ages[5] - sequence_ages[4] = 40 - 35 = 5
age = 5: sequence_ages[6] - sequence_ages[5] = 50 - 40 = 10
